```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_kmeans=pd.read_csv(r'K:\machine learning\kmeans.csv')
df_kmeans.head()

```

```python
df_kmeans = df_kmeans[["Annual_Income", "Spending_Score"]]
df_kmeans.head()
plt.scatter(df_kmeans["Annual_Income"],df_kmeans["Spending_Score"])
```

# Scale the model

```python
from sklearn.preprocessing import StandardScaler
scale_obj=StandardScaler()
scaled_data=scale_obj.fit_transform(df_kmeans)
```

```python
from sklearn.cluster import KMeans
kmeans_obj=KMeans(n_clusters=5,random_state=56,init='k-means++')
k_model=kmeans_obj.fit_predict(scaled_data)
```

```python
df_kmeans["Clusters"]=k_model
```

```python
df_kmeans
```

```python
df_2=df_kmeans[["Annual_Income","Spending_Score"]]
df_2.head()
```

```python
result=[]
for i in range(1,10):
    k_obj_2=KMeans(n_clusters=i,random_state=56,init='k-means++')
    k_obj_2.fit(scaled_data)
    result.append(k_obj_2.inertia_)
```

```python
print(result)
```

```python
#finding Good cluster number by using L bow method
plt.plot(range(1,10),result,marker='*')
```

```python
k_ob_3=KMeans(n_clusters=4,random_state=56,init='k-means++')
model=k_ob_3.fit_predict(scaled_data)
```

```python

df_2["cluster"]=model
df_2.head()
```

```python
for i in df_2.cluster.unique():
    plt.scatter(df_2[df_2["cluster"]==i].Annual_Income,df_2[df_2["cluster"]==i].Spending_Score,label=f"Clusters{i}")
plt.legend()
plt.show()
```
